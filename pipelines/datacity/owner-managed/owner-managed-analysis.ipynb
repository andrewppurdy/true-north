```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('companieslist_financialsColumnLayout_.csv')
```

```python
growth = data['BestEstimateGrowthPercentagePerYear']
incorporation_date = data['SortableIncorporationDate']
timestamp = pd.to_datetime(incorporation_date, format=f'%Y-%m-%d').astype(int).div(10**9).astype(int).div((86400*365.25)).add(1970).round(2)
frame = pd.DataFrame(data={'growth': growth, 'date':timestamp})
frame_cut = frame[frame['date'] >= 2005]
```

```python
plt.scatter(frame_cut['date'], frame_cut['growth'], marker='x', alpha=0.5)
plt.show()
```
